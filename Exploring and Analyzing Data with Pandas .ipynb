## Task 1: Load the Data
Objective: Load the dataset into a Pandas DataFrame.

Instructions:

Create a DataFrame from the provided dataset.

Display the first few rows of the dataset to ensure it loaded correctly.

```python
import pandas as pd

# Create a basic dataset
data = {
    'employee_id': range(1, 21),
    'name': ['John', 'Jane', 'Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack', 'Karen', 'Leo', 'Mia', 'Nina', 'Olivia', 'Paul', 'Quinn', 'Rita'],
    'age': [22, 25, 28, 35, 40, 42, 50, 60, 65, 30, 28, 24, 37, 41, 48, 33, 29, 32, 45, 38],
    'department': ['HR', 'Engineering', 'Marketing', 'Finance', 'Sales', 'HR', 'Engineering', 'Marketing', 'Finance', 'Sales', 'HR', 'Engineering', 'Marketing', 'Finance', 'Sales', 'HR', 'Engineering', 'Marketing', 'Finance', 'Sales'],
    'salary': [50000, 60000, 55000, 75000, 80000, 85000, 100000, 110000, 120000, 65000, 67000, 72000, 85000, 90000, 95000, 48000, 67000, 70000, 80000, 88000],
    'work_experience': [1, 3, 5, 7, 10, 12, 15, 18, 20, 3, 1, 6, 4, 8, 5, 7, 3, 2, 6, 5],
}


# Creating dataframe
df = pd.DataFrame(data)

# Display few rows from dataset
display(df.head())
```

### Task 2: Data Exploration

Objective: Explore the dataset to get an understanding of its structure.

Instructions:

Check for missing values in the dataset.

Get the summary statistics of numeric columns (age, salary, work_experience).

```python
## Finding missing values in dataset
missing_values = df.isnull()
display(missing_values)

```

```python
# finding summary statistics (mean. median, std, min, and max) of numeric columns (age, salary, work_experience).
age_mean = df['age'].mean() 
age_median = df['age'].median()
age_std = df['age'].std()
age_min = df['age'].min()
age_max = df['age'].max()

salary_mean = df['salary'].mean()
salary_median = df['salary'].median()
salary_std = df['salary'].std()
salary_min = df['salary'].min()
salary_max = df['salary'].max()

work_experience_mean = df['work_experience'].mean()
work_experience_median = df['work_experience'].median()
work_experience_std = df['work_experience'].std()
work_experience_min = df['work_experience'].min()
work_experience_max = df['work_experience'].max()

display(f"Age - Average: {age_mean}, Median: {age_median}, STD: {age_std}, Minimum: {age_min}, Maximum: {age_max}")
display(f"Salary - Average: {salary_mean}, Median: {salary_median}, STD: {salary_std}, Minimum: {salary_min}, Maximum: {salary_max}")
display(f"Work Experience - Mean: {work_experience_mean}, Median: {work_experience_median}, STD: {work_experience_std}, Minimum: {work_experience_min}, Maximum: {work_experience_max}")
```

### Task 3: Data Transformation

Objective: Perform basic data transformations such as creating new columns and renaming existing ones.

Instructions:

Create a new column called years_to_retirement by subtracting the age from 65.

Rename the salary column to annual_salary.

Sort the DataFrame based on work_experience in descending order.

```python
##Creating new column named 'years_to_retirement'
df['years_to_retirement'] = 65 - df['age']
display(df)
```

```python
## rename salary column
df.rename(columns={'salary':'annual_salary'}, inplace=True)
df
```

```python
## Sorting DataFrame based on work_experience in descending order.
df_sorted = df.sort_values(by='work_experience', ascending=False)
df_sorted
```

### Task 4: Handling Missing Data

Objective: Handle missing data (if any).

Instructions:

Introduce some missing values in the annual_salary column.

Handle the missing data by either filling it with the mean salary or dropping the rows with missing values.

```python
##Introducing missing values to dataset
df.loc[7, 'annual_salary'] = None
df.loc[2, 'annual_salary'] = None

display(df)
```

```python
df_dropped_row = df.dropna()   #drop rows with missing value in dataset
df_dropped_row
```

```python
display(len(df_dropped_row)) # display count of remaining rows in dataset
```

## Task 5: Data Filtering and Selection
Objective: Select specific data based on conditions.

Instructions:

Filter the data to show only employees who are older than 30.

Select only the columns name, age, and annual_salary.

```python
##  filter employees who are older than 30.
filtered_df = df[df['age'] > 30]
filtered_df
```

```python
# select and display columns name, age, and annual_salary.
df_selected_column = df[['name', 'age', 'annual_salary']]
df_selected_column
```

### Task 6: Data Aggregation
Objective: Perform group-based aggregation.

Instructions:

Group the data by department and calculate the average salary for each department.

```python
## Group by department and display with average salary of each department
grouped_department = df.groupby('department')['annual_salary'].mean().reset_index()
grouped_department
```
